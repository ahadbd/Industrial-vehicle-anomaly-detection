```python
import pandas as pd
import matplotlib.pyplot as plt

# Load anomaly results
df = pd.read_csv("../data/anomaly_results.csv")

# Separate normal vs anomalies
normal = df[df['anomaly'] == 1]
anomaly = df[df['anomaly'] == -1]

# Scatter plot: speed vs engine_rpm
plt.figure(figsize=(10,6))
plt.scatter(normal['speed'], normal['engine_rpm'], c='blue', label='Normal', alpha=0.6)
plt.scatter(anomaly['speed'], anomaly['engine_rpm'], c='red', label='Anomaly', alpha=0.8)
plt.xlabel('Speed (km/h)')
plt.ylabel('Engine RPM')
plt.title('Industrial Vehicle Anomaly Detection')
plt.legend()
plt.grid(True)
plt.show()


# Scatter Plot: Fuel Rate vs Idle Time

plt.figure(figsize=(10,6))
plt.scatter(normal['fuel_rate'], normal['idle_time'], c='blue', label='Normal', alpha=0.6)
plt.scatter(anomaly['fuel_rate'], anomaly['idle_time'], c='red', label='Anomaly', alpha=0.8)
plt.xlabel('Fuel Rate (L/h)')
plt.ylabel('Idle Time (s)')
plt.title('Industrial Vehicle Anomaly Detection: Fuel Rate vs Idle Time')
plt.legend()
plt.grid(True)
plt.show()

# Summary metrics
total_points = len(df)
anomaly_count = len(anomaly)
normal_count = len(normal)
anomaly_percentage = anomaly_count / total_points * 100

print(f"Total data points: {total_points}")
print(f"Normal points: {normal_count}")
print(f"Anomalies detected: {anomaly_count} ({anomaly_percentage:.2f}%)")
```
